```python
import os
import librosa
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv("../../KEMDy20_v1_1/Splitting/Train.csv")
test = pd.read_csv("../../KEMDy20_v1_1/Splitting/Test.csv")
```

```python
def get_feature(file_path: str, mfcc_len: int = 30, mean_signal_length: int = 100000):
    signal, fs = librosa.load(file_path)
    s_len = len(signal)

    if s_len < mean_signal_length:
        pad_len = mean_signal_length - s_len
        pad_tile = pad_len // s_len + 2
        signal = np.tile(signal, pad_tile)[:mean_signal_length]
    else:
        pad_len = s_len - mean_signal_length
        pad_len //= 2
        signal = signal[pad_len:pad_len + mean_signal_length]
        
    mfcc = librosa.feature.mfcc(y=signal, sr=fs, n_mfcc=mfcc_len)
    mfcc = mfcc.T
    
    feature = mfcc
    
    print(file_path)
    return feature
```

```python
train_feature_list = []
for path in train["Path"]:
    features = get_feature(path)
    train_feature_list.append(features)
```

```python
test_feature_list = []
for path in test["Path"]:
    features = get_feature(path)
    test_feature_list.append(features)
```

```python
train_array = np.stack(train_feature_list, axis=0)
test_array = np.stack(test_feature_list, axis=0)
```

```python
save_path = "../../KEMDy20_v1_1/Extract/"
if not os.path.exists(save_path):
    os.mkdir(save_path)
```

```python
np.savez(save_path+"Dataset_MFCC",
         train_x = train_array,
         train_y = train["Emotion"].values,
         test_x = test_array,
         test_y = test["Emotion"].values
         )
```
