```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
```

```python
train = pd.read_csv("../../KEMDy20_v1_1/Splitting/Train_Origin.csv")
test = pd.read_csv("../../KEMDy20_v1_1/Splitting/Test_Origin.csv")
```

```python
# 레이블 인코딩
le = LabelEncoder()
le.fit(train.Emotion)
encoded_train_y = le.transform(train.Emotion)

# train_y에서 사용된 LabelEncoder 객체를 그대로 사용하여 test_y 인코딩
encoded_test_y = le.transform(test.Emotion)
```

```python
import matplotlib.pyplot as plt

# 서브플롯 만들기
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(10, 20))

# 데이터
emotions = ['angry', 'disqust', 'fear', 'happy', 'neutral', 'sad', 'surprise']
colors = ['red', 'green', 'black', 'yellow', 'lightgray', 'blue', 'purple']

# 서브플롯에 그래프 그리기
for i, ax in enumerate(axes.flat):
    if i < len(emotions):
        data = train[train["Emotion"]==emotions[i]]
        ax.scatter(data["Valence"], data["Arousal"], c=colors[i])
        ax.set_xlim([1, 5])
        ax.set_ylim([1, 5])
        ax.axvline(x=3, color='red', linestyle='--')
        ax.axhline(y=3, color='red', linestyle='--')
        ax.set_title(emotions[i])

# 그래프 보이기
plt.tight_layout()
plt.show()
```
