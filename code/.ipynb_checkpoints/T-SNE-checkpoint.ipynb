```python
import numpy as np
tsne_dataset = np.load("../../KEMDy20_v1_1/tsne/feature_tsne.npz")
```

```python
x = tsne_dataset["x"]
y = tsne_dataset["y"]
```

```python
x = tsne_dataset["f1"]
y = tsne_dataset["f2"]

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
plt.scatter(x, y)
plt.colorbar()
plt.show()
```

```python
x = x.reshape(6312, 30)
y = y.reshape(6312, 7)
y= np.argmax(y, axis=1)
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

tsne = TSNE(n_components=2, random_state=0)
digits_tsne = tsne.fit_transform(x)
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(digits_tsne[:, 1], digits_tsne[:, 0], c=y, cmap=plt.cm.get_cmap('jet', 7))
plt.colorbar()
plt.show()
```
