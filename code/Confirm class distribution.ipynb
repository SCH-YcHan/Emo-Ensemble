```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_origin = pd.read_csv("../../KEMDy20_v1_1/Splitting/Train_Origin.csv")
test_origin = pd.read_csv("../../KEMDy20_v1_1/Splitting/Test_Origin.csv")

train = pd.read_csv("../../KEMDy20_v1_1/Splitting/Train.csv")
test = pd.read_csv("../../KEMDy20_v1_1/Splitting/Test.csv")
```

```python
emotion_colors = {'angry': 'red', 'disqust': 'green', 'fear': 'black',
                  'happy': 'yellow', 'neutral': 'lightgray', 'sad': 'blue', 'surprise': 'purple'}

train_origin_counts = train_origin['Emotion'].value_counts().sort_index()
test_origin_counts = test_origin['Emotion'].value_counts().sort_index()

train_counts = train['Emotion'].value_counts().sort_index()
test_counts = test['Emotion'].value_counts().sort_index()
```

```python
fig = plt.figure(figsize=(15, 7))

plt.subplot(1,2,1)
train_origin_counts.plot(kind='bar', color=[emotion_colors[e] for e in train_origin_counts.index])
plt.title('Distribution of Emotions (Train)')
plt.xlabel('Emotions')
plt.ylabel('Number of Occurrences')
for i, v in enumerate(train_origin_counts):
    plt.text(i, v + 100, str(v), color='black', fontsize=12, ha='center')

plt.subplot(1,2,2)
test_origin_counts.plot(kind='bar', color=[emotion_colors[e] for e in test_origin_counts.index])
plt.title('Distribution of Emotions (Test)')
plt.xlabel('Emotions')
#plt.ylabel('Number of Occurrences')
for i, v in enumerate(test_origin_counts):
    plt.text(i, v + 22, str(v), color='black', fontsize=12, ha='center')

plt.tight_layout()
plt.subplots_adjust(wspace=0.1)
#plt.show()
plt.savefig('../Class_distribution_origin.png', dpi=300)
```

```python
fig = plt.figure(figsize=(15, 7))

plt.subplot(1,2,1)
train_counts.plot(kind='bar', color=[emotion_colors[e] for e in train_counts.index])
plt.title('Distribution of Emotions (Train)')
plt.xlabel('Emotions')
plt.ylabel('Number of Occurrences')
for i, v in enumerate(train_counts):
    plt.text(i, v + 130, str(v), color='black', fontsize=12, ha='center')

plt.subplot(1,2,2)
test_counts.plot(kind='bar', color=[emotion_colors[e] for e in test_counts.index])
plt.title('Distribution of Emotions (Test)')
plt.xlabel('Emotions')
#plt.ylabel('Number of Occurrences')
for i, v in enumerate(test_counts):
    plt.text(i, v + 25, str(v), color='black', fontsize=12, ha='center')

plt.tight_layout()
plt.subplots_adjust(wspace=0.1)
#plt.show()
plt.savefig('../Class_distribution_add.png', dpi=300)
```
